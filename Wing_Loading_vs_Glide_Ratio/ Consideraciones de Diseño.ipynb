#Grafica Wing Loading vs Glide Ratio

La gr√°fica de **wing loading** (carga alar) vs **glide ratio** (relaci√≥n de planeo) es muy √∫til para analizar y comparar el rendimiento de diferentes parapentes, especialmente en t√©rminos de velocidad y eficiencia de planeo:

### ¬øPor qu√© es √∫til esta gr√°fica?

1. **Relaci√≥n entre carga y eficiencia de planeo**:
   - Al graficar **wing loading** (en el eje X) contra **glide ratio** (en el eje Y), puedes visualizar c√≥mo cambia la capacidad de planeo a medida que var√≠a la carga alar.
   - Un **mayor wing loading** (m√°s peso por unidad de √°rea de vela) normalmente resulta en un menor **glide ratio**, lo que significa que el parapente desciende m√°s r√°pido y cubre menos distancia horizontal.
   - Un **menor wing loading** tiende a producir un mayor glide ratio, lo que puede permitir que el parapente cubra una mayor distancia horizontal con una p√©rdida de altitud m√°s lenta.

2. **Selecci√≥n del mejor parapente**:
   - Para el sistema de recuperaci√≥n del CubeSat, es importante optimizar el **wing loading** para lograr un buen equilibrio entre la **velocidad de descenso** y la **distancia horizontal recorrida**.
   - Un glide ratio alto puede ser beneficioso si se quiere que el CubeSat aterrice suavemente y cubra cierta distancia controlada durante el descenso.
   - Un wing loading alto podr√≠a ser √∫til si se prefiere un descenso m√°s r√°pido y controlado.

3. **Ajuste del sistema de control PI**:
   - Entender la relaci√≥n entre el wing loading y el glide ratio tambi√©n te ayuda a ajustar tu sistema de control PI (con los servos) de manera m√°s precisa.
   - Un parapente con un mayor wing loading puede requerir ajustes m√°s sensibles en el control PI para mantener la estabilidad, mientras que un parapente con un menor wing loading podr√≠a ser m√°s f√°cil de controlar durante el descenso.

4. **An√°lisis de escenarios reales**:
   - Esta gr√°fica puede ayudarte a simular diferentes escenarios con tu CubeSat. Por ejemplo, si el CubeSat cae en condiciones de viento fuerte o turbulencias, puedes usar la gr√°fica para elegir un parapente con un wing loading m√°s alto que pueda ser m√°s estable en condiciones adversas, incluso si eso implica un menor glide ratio.

### Conclusi√≥n:
Esta gr√°fica te permitir√° hacer una selecci√≥n m√°s informada del parapente que mejor se ajuste a las necesidades de tu proyecto, ya sea que busques un descenso r√°pido o un control m√°s preciso durante el planeo.
```python
import matplotlib.pyplot as plt
import numpy as np

# Convertir las columnas a valores num√©ricos, forzando a NaN los valores no num√©ricos
all_sheets2["Wing Loading kg/m2"] = pd.to_numeric(all_sheets2["Wing Loading kg/m2"], errors='coerce')
all_sheets2["Glide Ratio"] = pd.to_numeric(all_sheets2["Glide Ratio"], errors='coerce')

# Eliminar filas con valores NaN en Wing Loading o Glide Ratio
all_sheets2.dropna(subset=["Wing Loading kg/m2", "Glide Ratio"], inplace=True)

# Asegurarse de que no haya NaN en la columna "Parapentes deportivos" y sea de tipo cadena
all_sheets2["Parapentes deportivos"] = all_sheets2["Parapentes deportivos"].fillna("Sin Nombre").astype(str)

# Crear una lista de colores para asignar un color distinto a cada parapente
colors = np.random.rand(len(all_sheets2))

# Crear la gr√°fica de dispersi√≥n (scatter plot)
plt.figure(figsize=(14, 8))

# Graficar Wing Loading kg/m2 en el eje X y Glide Ratio en el eje Y con colores aleatorios
plt.scatter(all_sheets2["Wing Loading kg/m2"], all_sheets2["Glide Ratio"], c=colors, cmap='viridis', s=50)

# Agregar el nombre de cada punto usando la columna "Parapentes deportivos"
#for i in range(len(all_sheets2)):
#    plt.text(all_sheets2["Wing Loading kg/m2"].iloc[i],
 #            all_sheets2["Glide Ratio"].iloc[i],
  #           all_sheets2["Parapentes deportivos"].iloc[i],
   #          fontsize=9, ha='right')

# Configurar etiquetas y t√≠tulo
plt.xlabel('Wing Loading (kg/m¬≤)')
plt.ylabel('Glide Ratio')
plt.title('Wing Loading vs Glide Ratio')

# Agregar cuadricula
plt.grid(True)

# Mostrar la gr√°fica con los puntos en diferentes colores
plt.show()
```
