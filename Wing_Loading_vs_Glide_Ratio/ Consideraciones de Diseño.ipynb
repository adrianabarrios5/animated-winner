#Grafica Wing Loading vs Glide Ratio

La gráfica de **wing loading** (carga alar) vs **glide ratio** (relación de planeo) es muy útil para analizar y comparar el rendimiento de diferentes parapentes, especialmente en términos de velocidad y eficiencia de planeo:

### ¿Por qué es útil esta gráfica?

1. **Relación entre carga y eficiencia de planeo**:
   - Al graficar **wing loading** (en el eje X) contra **glide ratio** (en el eje Y), puedes visualizar cómo cambia la capacidad de planeo a medida que varía la carga alar.
   - Un **mayor wing loading** (más peso por unidad de área de vela) normalmente resulta en un menor **glide ratio**, lo que significa que el parapente desciende más rápido y cubre menos distancia horizontal.
   - Un **menor wing loading** tiende a producir un mayor glide ratio, lo que puede permitir que el parapente cubra una mayor distancia horizontal con una pérdida de altitud más lenta.

2. **Selección del mejor parapente**:
   - Para el sistema de recuperación del CubeSat, es importante optimizar el **wing loading** para lograr un buen equilibrio entre la **velocidad de descenso** y la **distancia horizontal recorrida**.
   - Un glide ratio alto puede ser beneficioso si se quiere que el CubeSat aterrice suavemente y cubra cierta distancia controlada durante el descenso.
   - Un wing loading alto podría ser útil si se prefiere un descenso más rápido y controlado.

3. **Ajuste del sistema de control PI**:
   - Entender la relación entre el wing loading y el glide ratio también te ayuda a ajustar tu sistema de control PI (con los servos) de manera más precisa.
   - Un parapente con un mayor wing loading puede requerir ajustes más sensibles en el control PI para mantener la estabilidad, mientras que un parapente con un menor wing loading podría ser más fácil de controlar durante el descenso.

4. **Análisis de escenarios reales**:
   - Esta gráfica puede ayudarte a simular diferentes escenarios con tu CubeSat. Por ejemplo, si el CubeSat cae en condiciones de viento fuerte o turbulencias, puedes usar la gráfica para elegir un parapente con un wing loading más alto que pueda ser más estable en condiciones adversas, incluso si eso implica un menor glide ratio.

### Conclusión:
Esta gráfica te permitirá hacer una selección más informada del parapente que mejor se ajuste a las necesidades de tu proyecto, ya sea que busques un descenso rápido o un control más preciso durante el planeo. También te dará una mejor idea de cómo ajustar tu sistema de control PI en función del rendimiento del parapente que elijas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Convertir las columnas a valores numéricos, forzando a NaN los valores no numéricos
all_sheets2["Wing Loading kg/m2"] = pd.to_numeric(all_sheets2["Wing Loading kg/m2"], errors='coerce')
all_sheets2["Glide Ratio"] = pd.to_numeric(all_sheets2["Glide Ratio"], errors='coerce')

# Eliminar filas con valores NaN en Wing Loading o Glide Ratio
all_sheets2.dropna(subset=["Wing Loading kg/m2", "Glide Ratio"], inplace=True)

# Asegurarse de que no haya NaN en la columna "Parapentes deportivos" y sea de tipo cadena
all_sheets2["Parapentes deportivos"] = all_sheets2["Parapentes deportivos"].fillna("Sin Nombre").astype(str)

# Crear una lista de colores para asignar un color distinto a cada parapente
colors = np.random.rand(len(all_sheets2))

# Crear la gráfica de dispersión (scatter plot)
plt.figure(figsize=(14, 8))

# Graficar Wing Loading kg/m2 en el eje X y Glide Ratio en el eje Y con colores aleatorios
plt.scatter(all_sheets2["Wing Loading kg/m2"], all_sheets2["Glide Ratio"], c=colors, cmap='viridis', s=50)

# Agregar el nombre de cada punto usando la columna "Parapentes deportivos"
#for i in range(len(all_sheets2)):
#    plt.text(all_sheets2["Wing Loading kg/m2"].iloc[i],
 #            all_sheets2["Glide Ratio"].iloc[i],
  #           all_sheets2["Parapentes deportivos"].iloc[i],
   #          fontsize=9, ha='right')

# Configurar etiquetas y título
plt.xlabel('Wing Loading (kg/m²)')
plt.ylabel('Glide Ratio')
plt.title('Wing Loading vs Glide Ratio')

# Agregar cuadricula
plt.grid(True)

# Mostrar la gráfica con los puntos en diferentes colores
plt.show()
```

### Consideraciones clave para el diseño a escala más pequeña

Si decides basarte en uno de los parapentes recomendados y escalar el diseño a uno más pequeño para el CubeSat, hay varias características que deberías tener en cuenta:

Sin embargo, un parapente con demasiados cajones será más difícil de construir y manejar a escala pequeña. Requiere más precisión en la alineación y distribución de las fuerzas.
     - **Efecto en el diseño**: Si escalas un parapente con muchos cajones, tendrás que mantener la precisión en el control, lo que aumenta la carga de trabajo en los servos.

2. **Aspect Ratio (Relación de Aspecto)**:
   - **Alto aspect ratio**: Mejora la eficiencia en planeo (glide ratio), pero reduce la maniobrabilidad y aumenta la sensibilidad a las turbulencias. Un parapente con un alto aspect ratio será más difícil de maniobrar en una trayectoria helicoidal, especialmente a pequeña escala.
     - **Efecto en el diseño**: Escalar un diseño con un aspect ratio alto puede hacer que tu parapente sea difícil de controlar. Un parapente muy alargado podría requerir más control de los servos para mantener la trayectoria.
   - **Bajo aspect ratio**: Mejora la maniobrabilidad, lo que es crucial para trayectorias complejas como la helicoidal. Sin embargo, sacrifica la eficiencia en planeo, lo que podría hacer que el parapente pierda altitud rápidamente.
     - **Efecto en el diseño**: Si eliges un diseño con un bajo aspect ratio, podrías tener mejor maniobrabilidad, pero necesitarás más precisión en los servos para controlar el descenso y compensar la pérdida de altitud rápida.

3. **Wing Loading**:
   - **Alto wing loading**: Proporciona mayor estabilidad en vientos moderados o turbulencias, pero aumenta la velocidad de descenso. Esto podría hacer que el CubeSat descienda demasiado rápido y afecte el tiempo de vuelo.
     - **Efecto en el diseño**: Escalar un parapente con un wing loading alto puede dar mayor estabilidad, pero debes tener cuidado de no exceder los límites de tiempo de descenso.
   - **Bajo wing loading**: Mejora la maniobrabilidad y hace que el descenso sea más lento, lo que puede ayudarte a mantener la trayectoria helicoidal. Sin embargo, parapentes con un wing loading bajo pueden verse más afectados por el viento.
     - **Efecto en el diseño**: Si escalas un parapente con un wing loading bajo, necesitarás un buen sistema de control para compensar las posibles desviaciones por el viento.

4. **Span (Envergadura)**:
   - **Mayor span**: Un span mayor ayuda a mejorar el glide ratio, lo que podría ser beneficioso para mantener el planeo por más tiempo, pero un parapente más grande será menos maniobrable.
     - **Efecto en el diseño**: Si reduces demasiado la envergadura al escalar, podrías perder eficiencia en planeo. Idealmente, mantén un balance entre el tamaño y la eficiencia del planeo.
   - **Menor span**: Mejora la maniobrabilidad pero puede reducir la eficiencia del planeo. Si la envergadura es demasiado pequeña, podrías no cumplir con el tiempo de descenso estimado.
     - **Efecto en el diseño**: Un menor span facilita la maniobrabilidad, lo cual es útil para la trayectoria helicoidal, pero puede reducir la distancia de planeo.

5. **Max Chord (Cuerda Máxima)**:
   - **Mayor max chord**: Mejora la estabilidad aerodinámica, lo cual es útil para mantener la trayectoria helicoidal en condiciones de viento.
     - **Efecto en el diseño**: Si eliges un parapente con una cuerda mayor, es probable que el control sea más estable, pero podría dificultar la maniobrabilidad en giros cerrados.
   - **Menor max chord**: Mejora la maniobrabilidad, pero reduce la estabilidad, lo cual puede afectar el control preciso en la trayectoria helicoidal.
     - **Efecto en el diseño**: Un diseño con menor cuerda sería más ágil en giros, pero podría perder estabilidad en la trayectoria de descenso.

### Factores adicionales que podrías estar olvidando:
- **Control de los servos**: Asegúrate de que el sistema de control que estás utilizando pueda manejar los ajustes necesarios para mantener una trayectoria helicoidal precisa, especialmente si escalas el diseño.
- **Tamaño de las líneas de suspensión**: Al reducir la escala del parapente, asegúrate de que las líneas de suspensión sean lo suficientemente fuertes para soportar el peso del CubeSat, pero no demasiado pesadas como para afectar negativamente el vuelo.
- **Condiciones meteorológicas específicas**: El parapente debe ser optimizado para las condiciones de viento y temperatura en São Paulo. Un parapente más estable será más adecuado si hay turbulencias o vientos fuertes en la zona de aterrizaje.

¿Te gustaría discutir más sobre cómo ajustar los servos para el control en función del diseño? Esto será crucial para mantener la trayectoria helicoidal en tu parapente más pequeño.

Para que tu parapente mantenga una **trayectoria helicoidal controlada** durante el descenso del CubeSat, el sistema de control con **dos servos** y un **control proporcional-integral (PI)** será fundamental. Aquí te explico los aspectos clave para lograrlo y lo que debes considerar al implementar el sistema de control:

### 1. **Objetivos del Control del Parapente**:
   - **Mantener una trayectoria helicoidal**: Esto significa que el parapente debe realizar giros continuos mientras desciende, con un radio constante o controlado.
   - **Ajustar la velocidad de descenso**: El control debe regular la altura y la trayectoria de manera que el CubeSat no descienda demasiado rápido y cumpla con los 15 minutos de tiempo de descenso.
   - **Compensar perturbaciones**: Las condiciones atmosféricas, como vientos laterales o turbulencias, pueden desviar el parapente de la trayectoria deseada. El control PI debe corregir estos desvíos.

### 2. **Uso de los Servos**:
   - **Distribución del control**: Con dos servos, cada uno actuará sobre las **líneas de suspensión** o risers del parapente, típicamente controlando los frenos de cada lado (izquierdo y derecho). Al **tirar de un lado**, se inclina el parapente y gira hacia ese lado.
   - **Ajustes pequeños y continuos**: El control PI proporcionará ajustes pequeños pero continuos en la posición de los servos, de modo que el parapente pueda realizar giros suaves y mantenga la trayectoria helicoidal.
   
   **Relación entre los servos y el control**:
   - **Servo izquierdo**: Tirará del freno izquierdo, haciendo que el parapente gire hacia la izquierda. La fuerza aplicada por el servo determinará el radio del giro.
   - **Servo derecho**: Hará lo mismo en el lado derecho.
   - **Ambos servos en conjunto**: Al tirar ambos servos de manera controlada, puedes ajustar la inclinación del parapente para compensar cualquier desviación debida al viento.

### 3. **Control Proporcional-Integral (PI)**:
   - **Control Proporcional (P)**: Esta parte del controlador ajusta la posición de los servos en proporción al **error** entre la trayectoria real y la deseada. Si el parapente se desvía demasiado de la trayectoria helicoidal, el control P aplicará una mayor fuerza para corregir el error rápidamente.
   - **Control Integral (I)**: Este componente acumula los errores pasados y ajusta los servos para corregir desviaciones persistentes. Esto es crucial para asegurar que el parapente no se desvíe gradualmente de la trayectoria helicoidal por efectos del viento o turbulencia.
   - **Ajuste del PI**: Deberás sintonizar los parámetros de P e I de manera que no haya una **sobrecorrección** (lo que podría causar inestabilidad) ni una **corrección insuficiente** (lo que haría que el parapente se desvíe lentamente de su trayectoria).

### 4. **Factores que Afectan el Control de los Servos**:
   - **Tamaño del parapente**: Como el span será relativamente pequeño (0.8-1.2 metros), las correcciones de los servos tendrán un impacto mayor en el control de la trayectoria. Por lo tanto, los ajustes que hagas en los servos deben ser precisos para evitar desviaciones bruscas.
   - **Tensión en las líneas de suspensión**: Asegúrate de que los servos puedan generar la fuerza suficiente para mantener tensas las líneas, de modo que las correcciones tengan un efecto inmediato en la inclinación del parapente. Tus servos de 3.5 kg de fuerza parecen adecuados para este propósito.
   - **Velocidad de respuesta de los servos**: Los servos deben tener una **respuesta rápida** para corregir las desviaciones de la trayectoria helicoidal sin retrasos. Asegúrate de que los servos no tengan un tiempo de respuesta demasiado largo.

### 5. **Estrategia de Control para Mantener la Trayectoria Helicoidal**:
   - **Desviación angular constante**: Para mantener la trayectoria helicoidal, necesitas programar un ángulo de giro constante. El sistema PI ajustará continuamente los servos para asegurar que el parapente mantenga ese ángulo de giro, compensando el viento o las turbulencias.
   - **Control de la altura**: Además del ángulo de giro, deberás monitorear la altura en todo momento. Si el parapente empieza a descender demasiado rápido, el sistema PI deberá ajustar los frenos para aumentar la resistencia y reducir la velocidad de descenso.
   
   **Estrategia de control**:
   - **Controlar el giro**: Programar los servos para que tiren de las líneas de manera alternada (izquierda-derecha) para mantener el giro constante.
   - **Compensar la altura**: Si el parapente está descendiendo demasiado rápido, ambos servos deberían aplicar fuerza a los frenos simultáneamente para aumentar la resistencia.

### 6. **Condiciones Meteorológicas y su Impacto en el Control**:
   - **Viento lateral**: Podría desviar el parapente de su trayectoria helicoidal. El control PI deberá ajustarse para compensar el viento, aumentando la fuerza aplicada por uno de los servos para contrarrestar el empuje del viento.
   - **Turbulencias**: Podrían causar movimientos bruscos en el parapente, lo que afectará su estabilidad. Asegúrate de que el PI esté ajustado para corregir rápidamente las perturbaciones sin sobrecompensar, lo que podría causar oscilaciones.

### 7. **Simulación y Pruebas**:
   - **Simulación previa**: Antes de probar el parapente en la vida real, sería ideal realizar simulaciones del sistema de control en un software como **MATLAB** o **Simulink**. Simula diferentes velocidades de viento y perturbaciones para ajustar el PI de manera precisa.
   - **Pruebas con prototipos pequeños**: Puedes construir una versión pequeña del sistema de control y realizar pruebas en un entorno controlado (como en interiores o en condiciones de viento ligero). Esto te permitirá ajustar los parámetros del PI y comprobar la respuesta de los servos antes de la prueba final.
